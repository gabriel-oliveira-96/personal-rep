```python
import pandas as pd
```

```python
tbl_postos = pd.read_csv('tbl_postos.csv', sep=',')
tbl_postos.columns = ['num_posto','num_ssis','nom_ssis','nom_posto','bin_ena','val_product']
tbl_postos.head()
```

```python
tbl_vazao = pd.read_csv('tbl_vazao.csv', sep=',')
tbl_vazao.columns = ['num_posto','dat_medicao','val_vaz_natr']
tbl_vazao.head()
```

```python
tabela_total = pd.merge(tbl_postos, tbl_vazao, on=["num_posto", "num_posto"])
tabela_total = tabela_total.query('bin_ena == 1')
tabela_total['ENA'] = tabela_total['val_product'] * tabela_total['val_vaz_natr']
tabela_total.head()
```

```python
ssis_agg = tabela_total[['nom_ssis','ENA']]
```

```python
ENA_agg_ssis = ssis_agg.groupby("nom_ssis")["ENA"].sum()
ENA_agg_ssis = pd.DataFrame(ENA_agg_ssis)
pd.set_option('float_format', '{:f}'.format)
```

```python
ENA_agg_ssis
```

```python
daily_avg = ssis_agg.groupby("nom_ssis")["ENA"].mean()
daily_avg = pd.DataFrame(daily_avg)
daily_avg
```

```python
ssis_agg2 = tabela_total[['nom_ssis','dat_medicao','ENA']]
ssis_agg2.head()
```

```python
ssis_agg2['dat_medicao'] = ssis_agg2['dat_medicao'].apply(pd.to_datetime)
```

```python
ssis_agg2['mes_medicao'] = ssis_agg2['dat_medicao'].dt.month
ssis_agg2.head()
```

```python
dmonthy_avg = ssis_agg2.groupby(['nom_ssis','mes_medicao'])["ENA"].mean()
dmonthy_avg = pd.DataFrame(dmonthy_avg)
dmonthy_avg
```
